```python
import csv
import pandas as pd
from datetime import datetime
import calendar
import warnings
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
warnings.filterwarnings("ignore")
```

```python
try:
    df = pd.read_csv('../CSV/mdc.csv', encoding='utf-8')
except UnicodeDecodeError:
    try:
        df = pd.read_csv('../CSV/mdc.csv', encoding='latin1')
    except UnicodeDecodeError:
        try:
            df = pd.read_csv('../CSV/mdc.csv', encoding='ISO-8859-1')
        except UnicodeDecodeError:
            df = pd.read_csv('../CSV/mdc.csv', encoding='utf-16')
```

```python
df = df[['title','entity','year','mpa_rating','imdb_rating','tomato_meter','imdb_gross']]
df.head()
```

```python
def format_as_dollars(imdb_gross):
    return '${:,.2f}'.format(imdb_gross)
df['imdb_gross'] = df['imdb_gross'].apply(format_as_dollars)
```

```python
marvel_df = df[df['entity'] == 'MARVEL']
marvel_df = marvel_df.reset_index(drop=True)
dc_df = df[df['entity'] == 'DC']
dc_df = dc_df.reset_index(drop=True)
dc_df
```

```python
marvel_df.to_csv('../CSV/marvel_movies.csv')
```

```python
dc_df.to_csv('../CSV/dc_movies.csv')
```
