```python
import csv
import pandas as pd
```

```python
df = pd.read_csv("../CSV/comic_characters.csv")
df.head()
```

```python
marvel_df = df[df["Universe"] == "Marvel"]
marvel_df.head()
```

```python
dc_df = df[df["Universe"] == "DC"]
dc_df.head()
```

```python
marvel_good_df = marvel_df[marvel_df["Alignment"] == "Good"]
marvel_good_df.head()
```

```python
marvel_bad_df = marvel_df[marvel_df["Alignment"] == "Bad"]
marvel_bad_df.head()
```

```python
appearances_bins = [0,10,100,500,1000,2000,3000,50000]
group_names = ["<10","<100","<500","<1000","<2000","<3000",">3000"]

marvel_df["Appearance Bins"] = pd.cut(marvel_df["Appearances"],
                                     appearances_bins, labels=group_names,
                                      include_lowest=True)
marvel_df.head()
```
